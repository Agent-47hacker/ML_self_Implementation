```python
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target
x.shape,y.shape
```

```python
x_ones  = np.ones((x.shape[0],1))
x = np.append(x,x_ones ,axis = 1)
print(x.shape)
x
```

```python
def sigmoid(x,m):
    z = (np.dot(x,m))
    return 1 / (1 + np.exp(-z))
```

```python
def sigmoid_r(z): 
    res = 1 / (1 + np.exp(-z))
    return res
```

```python
def sgd(x,y,lr,n,feature_count,m):
    slope = np.zeros((feature_count))
    #slope[feature_count - 1] = 1
    for i in range(n):
        t = 0
        for j in range(feature_count):
            t += m[j]*x[i][j]
        r = sigmoid_r(t)
        for j in range(feature_count):
            slope[j] +=(-1/n)* (y[i] - r)*x[i][j]
    m = m - lr*slope
    return m
```

```python
def cost(x,y,m,n,feature_count=0):
    cost = 0
    for i in range(n):
        t = sigmoid(x[i],m)
        cost += (1/n) * (-y[i] * np.log(t)-(1-y[i])*np.log(1-t))
    return cost
# def cost(x,y,m,n):
#     prob = sigmoid_r(x,m)
#     print("data point:",prob)
#     return (-y * np.log(prob) - (1 - y) * np.log(1 - prob)).mean()
```

```python
def fit(x,y,num_it,lr,n,feature_count):
    m = np.zeros(feature_count)
    for i in range(num_it):
        m = sgd(x,y,lr,n,feature_count,m)
        print(m)
        print(i , "cost:",cost(x,y,m,n))
    return m
```

```python
n = x.shape[0]
feature_count = x.shape[1]
m = fit(x,y,2000,.00001,n,feature_count)
```

```python
predictions = []
for i in range(len(x)):
    z = (np.dot(x[i],m))
    sig = 1 / (1 + np.exp(-z))
    predictions.append(np.round(sig))
```

```python
c = [predictions == y]
c = np.sum(c)
accuracy = c/len(y)
accuracy
```
