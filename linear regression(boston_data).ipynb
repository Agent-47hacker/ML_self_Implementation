```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = np.genfromtxt("training_boston_x_y_train.csv" ,delimiter = ',')
print(data.shape)
x= data[:,:13]
# x = x.reshape(-1,13)
y=data[:,13:14]
y = y.ravel()
print(x.shape)
print(y.shape)
a = np.ones((x.shape[0],1))
x = np.append(x,a,axis = 1)
print(x.shape)
print(x[1])
```

```python
def gd(x,y,lr,num,feature_count,m):
    slope = np.zeros(feature_count)
    slope[feature_count-1] = 1
    for i in range(num):
        t=0
        for j in range(feature_count):
            t += m[j]*x[i][j]
        for j in range(feature_count):
            slope[j] = (-2/num)*(y[i] -t)*x[i][j]
        m = m - lr*slope
    return m
```

```python
def cost(x,y,m,num,feature_count):
    cost = 0
    for i in range(num):
        cost += (1/num)*(y[i]- (m*x[i]).sum())**2
    return(cost)
```

```python
def fit (x,y,lr,epochs,num,feature_count):
    m = np.zeros(feature_count)
    print('fit',m)
    for i in range(epochs):
        m = gd(x,y,lr,num,feature_count,m)
        print(m)
        print(i,'cost',cost(x,y,m,num,feature_count))
    return m
```

```python
num = x.shape[0]
feature_count = x.shape[1]
print(feature_count)
m = fit(x,y,0.1,1000,num,feature_count)
```

```python
pred_y=np.zeros(379)
for i in range (num):
    l=0
    for j in range(feature_count):
        l+= x[i][j]*m[j]
    pred_y[i]=l
print(pred_y)
```
